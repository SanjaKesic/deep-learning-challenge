```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
columns_to_drop = ['EIN','NAME']
application_df = application_df.drop(columns=columns_to_drop)
# Print the resulting DataFrame
application_df
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
unique_value_counts = application_df.nunique()
unique_value_counts
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
cutoff_value = 10
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = list(application_type_counts[application_type_counts < cutoff_value].index)


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
classification_type_counts = application_df['CLASSIFICATION'].value_counts()
classification_type_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
classification_type_counts = classification_type_counts[classification_type_counts > 1]
classification_type_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
cutoff_value = 100
classification_type_counts = application_df['CLASSIFICATION'].value_counts()

classifications_to_replace = list(classification_type_counts[classification_type_counts < cutoff_value].index)
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
application_df_encoded = pd.get_dummies(application_df)
application_df_encoded
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = application_df_encoded['IS_SUCCESSFUL'].values
X = application_df_encoded.drop(columns='IS_SUCCESSFUL').values
# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

number_input_features = len(X_train[0])
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, activation="relu", input_dim=number_input_features))
# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn.fit(X_train_scaled, y_train, epochs=500)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

**Overview of the analysis:**

The goal of the analysis is to help the nonprofit foundation Alphabet Soup select the applicants for funding with the best chance of success in their ventures. A binary classifier was created in order to predict whether applicants would be successful if funded by Alphabet Soup.

There are more than 34,000 organizations that have received funding from Alphabet Soup over the years, and we are looking into that dataset.  The columns that capture metadata about each organization are as follows:

* **EIN** and **NAME**&mdash;Identification columns
* **APPLICATION_TYPE**&mdash;Alphabet Soup application type
* **AFFILIATION**&mdash;Affiliated sector of industry
* **CLASSIFICATION**&mdash;Government organization classification
* **USE_CASE**&mdash;Use case for funding
* **ORGANIZATION**&mdash;Organization type
* **STATUS**&mdash;Active status
* **INCOME_AMT**&mdash;Income classification
* **SPECIAL_CONSIDERATIONS**&mdash;Special considerations for application
* **ASK_AMT**&mdash;Funding amount requested
* **IS_SUCCESSFUL**&mdash;Was the money used effectively

The steps in this analysis were data preprocessing and compiling,training and evaluating the model.

* Data preprocessing
  * we took the column "IS_SUCCESSFUL" as the target - it is key data that tells us whether the money was used effectively
  * the features of the model are: application type, affiliation, classification, use_case, organization, status, income_amt, special_considerations and ask_amt-funding amount requested
  * EIN and NAME were removed as columns-variables since as identification columns they do not contribute to the quality of the analysis

* Compiling, Training, and Evaluating the Model
  * The initial neural network model had two layers with 80 neurons in the first layer and 30 neurons in the second layer, with 100 epochs. The accuracy achieved was 0.0.7258309125900269. The goal was to optimize the model to achieve accuracy of at least 75%
  * Steps taken were:
  1) reduction of the no of bins in preprocessing stage for the application type - result was acc of 72.4%.
  2) increase of the no of bins in preprocessing stage for the application type AND for the classification type by changing cutoff value - result was 72.59% accuracy
  3) increasing number of neurons in layer 1 from 80 to 90 - result was 72.54% accuracy
  4) increasing no of epochs from 100 to 500 - result was 72.56% accuracy

Summary - Overall, the steps taken to optimize the model did not achieve significantly higher accuracy. There was also an option to reduce the number of columns - variables to streamline the model, however, I decided not to take the steps since I am not entirely familiar with the meaning of codes and types in the dataset and I did not want any important information to be lost. In further optimization steps the model can be changed from Relu to another model.
